# **<font color=purple>List of IPL Season</font>**

```python
ipl['Season'].unique()
```

```python
plt.figure(figsize=(18,15))
sns.countplot(data=ipl,
              x='Season',
              order = ipl['Season'].value_counts().index.sort_values(),
              palette='cividis'
              )
c=ipl['Season'].value_counts()
plt.yticks(c)
plt.xticks()
plt.title('List of IPL Season',fontsize=16)
plt.show()
```

## **Onservation**<br>- We see that dataset contains data of matches played from season 2008-2019.

# **<font color=purple>Number of Match Played in each IPL Season</font>**

```python
plt.figure(figsize=(15,7))
sns.countplot(x='Season',
              data=ipl,
              order = ipl['Season'].value_counts().index.sort_values(),
              palette='cividis'
              )
plt.xlabel('Season',fontsize=12)
plt.ylabel('No: of matches',fontsize=12)
plt.title('Number of matches played in each IPL season',fontsize=16)
plt.show()
```

## **Observation**<br>- We see that Most number of matches were played in IPL-2013 Season

# **<font color=purple>Most IPL Matches played in a Venue</font>**

```python
ipl.groupby(['venue','id']).count().droplevel(level=1).index.value_counts()
```

```python
plt.figure(figsize = (12,8))
sns.countplot(y = 'venue',
              data = ipl,
              order = ipl['venue'].value_counts().iloc[:10].index,
              palette='gist_heat')
plt.xlabel('No: of matches',fontsize=12)
plt.ylabel('Venue',fontsize=12)
plt.title('Number of matches played in each Stadium',fontsize=16)
```

# **<font color=purple>Team has most winning matches</font>**

```python
ipl['winner'].value_counts().sum()
```

```python
ipl['winner'].value_counts()
```

```python
# Matches won by the Teams
plt.figure(figsize = (11,9))
sns.countplot(y = 'winner',
              data = ipl,
              order= ipl['winner'].value_counts().index,
              palette = 'copper' )
plt.xlabel('No: of Wins',fontsize=12)
plt.ylabel('Team',fontsize=12)
plt.title('Matches won by the Teams',fontsize=16)
plt.show()
```

## **Observation**<br>- We see that Mumbai Indians won most number of matches followed by Chennai Super Kings

# **<font color='purple'> Team has Most Toss Wins in IPL</font>** 

```python
plt.figure(figsize=(10,8))
sns.countplot(y='toss_winner',
              data=ipl,
              order=ipl['toss_winner'].value_counts().index,
              palette='copper')
plt.xlabel('Matches won by the Teams',fontsize=12)
plt.ylabel('Team',fontsize=12)
plt.title('Most of Toss Winning Team',fontsize=16)
plt.show()
```

## **Observation**<br>- We see that Mumbai Indians won most number of Tosses followed by Kolkata Knight Riders

# **<font color=purple>Toss Decision</font>**

```python
plt.figure(figsize=(6,6))
sizes = ipl.toss_decision.value_counts()
labels = ipl.toss_decision.value_counts().index
plt.pie(sizes,colors = ['#ff9999','#66b3ff'],
         labels=labels,
         autopct='%1.1f%%',
         startangle=90,
         pctdistance=0.75,
         explode = (0.025,0.025))

#draw white circle
centre_circle = plt.Circle((0,0),0.60,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.show()
```

## **Observation**<br>- We see that In more than 61% of matches, teams chose to field after winning the toss.

# **<font color=purple>Toss Decision to field or bat in each IPL season</font>**

```python
plt.figure(figsize=(15,7))
sns.countplot(x='Season',
              hue='toss_decision',
              data=ipl,
              order = ipl['Season'].value_counts().index.sort_values(),
              palette='deep')
plt.xlabel('Season',fontsize=12)
plt.ylabel('Count',fontsize=12)
plt.title('Decision to field or bat in each IPL season', fontsize=16)
plt.show()
```

## **Observation**<br>- We can see a trend that, decisions to field are more from IPL-2016 Season.

# **<font color=purple>Player of the Match</font>**

```python
plt.figure(figsize=(15,7))
player_of_match=ipl['player_of_match'].value_counts()[:10]
sns.barplot(player_of_match.index,
            player_of_match.values,
            palette='cividis')

plt.title("Player of the Match",fontsize=15)
plt.xlabel('Player',fontsize=12)
plt.ylabel('Count',fontsize=12)
plt.show()
```

## **Observation**<br>- We see that Chris Gayle won most Man of the Match Awards



# **<font color=purple>Mumbai Indians Team Winning chance after choosing field.</font>**

```python
mi=ipl[(ipl['toss_winner']=='Mumbai Indians')&(ipl['toss_decision']=='field')]
mi
```

```python
plt.figure(figsize=(10,8))
mi['winner'].value_counts().plot(kind='pie',autopct='%.2f%%')
plt.title('Teams Winning Chances after choosing Field')
plt.xlabel('Teams',fontsize=12)
plt.ylabel('chances',fontsize=12)
plt.show()
```

## **Observation**<br>- We see that there are 57.41% of Chances to Win if Mumbai Indians Choose Field.
