# **<font color=purple>Encoding Categorical Data</font>**

```python
#checking datatypes of each column
df.dtypes
```

### <font color=blue>Seperate the numerical data(int and float) and categorical data (object)</font>

```python
#creating first dataframe df_cat which hold object type data means categorical data
df_cat = df.select_dtypes(object) #select_dtypes() inbuilt method of dataframe
```

```python
#checking datatypes of df_cat
df_cat.dtypes
```

```python
#creating second dataframe df_num which hold numerical data (int and float)
df_num=df.select_dtypes(['int64','float64'])
```

```python
#checking datatypes of df_num
df_num.dtypes
```

### <font color=blue>Convert all Categorical Data into Numerical Data</font>

```python
for col in df_cat:
    #create object of LabelEncoder class
    le=LabelEncoder()
    df_cat[col]=le.fit_transform(df_cat[[col]])
```

```python
#check
df_cat.head()
```

```python
#checking datatypes of df_cat dataframe
df_cat.dtypes
```

## **Observation**<br>- We see all the columns of our data set are free from categorical datatype

#  **<font color=purple>Handling Skewness</font>**

### <font color=green>Check Skewness on each numerical column<font>

```python
#checking skewness of each numerical column
for col in df_num:
    s = df_num[col].skew()
    print("Column Name : {}, Skewness : {} ".format(col,s))
    sns.histplot(df_num[col])
    plt.show()
    sns.distplot(df_num[col])
    plt.show()
```

```python
for col in df_num:
    s = df_num[col].skew()
    print("{} : {} ".format(col,s))
```
